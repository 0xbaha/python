```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,5,11)
```

```python
x
```

```python
y = x ** 2
```

```python
y
```

```python
plt.plot(x,y,'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Data Contoh')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'b*-')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y,'b')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Contoh Judul')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,1,2])

axes.plot(x,y,'b')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Contoh Judul')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'b')
axes1.set_xlabel('X utama')
axes1.set_ylabel('Y utama')
axes1.set_title('Plot utama')

axes2.plot(x,y,'r')
axes2.set_xlabel('X dalam')
axes2.set_ylabel('Y dalam')
axes2.set_title('Plot dalam')
```

```python
fig = plt.figure(figsize=(8,4),dpi=100)
```

```python
fig,axes = plt.subplots(figsize=(8,4),dpi=100)

axes.plot(x,y,'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Plot judul')
```

```python
fig,axes = plt.subplots(figsize=(5,4),dpi=100)

axes.plot(x,y,'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Plot judul')

fig.savefig("contoh.png",dpi=100)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x ** 2, label="Data x pangkat 2")
axes.plot(x, x ** 3, label="Data x pangkat 3")

axes.legend(loc=4)
```
